```python
import json
import re
```

```python
with open('lyrics_lines.txt') as f:
  data = f.read()
  
print("Data type before reconstruction : ", type(data)) 
      
# reconstructing the data as a dictionary 
raw_lyrics = json.loads(data) 
print("Data type after reconstruction : ", type(raw_lyrics)) 
```

```python
def clean_title(s):
  # Replace non-breaking space (\xa0) with a regular space
  s = s.replace('\xa0', ' ')
    
  # Use a regular expression to remove or replace other unwanted characters
  s = re.sub(r'[^\x00-\x7F]+', '', s)  # Removes non-ASCII characters
  return s
def clean_string(l):
  s = clean_title(l)
  
  # reg expr to get rid of all text enclosed by **
  s = re.sub(r'\*[^*]*\*', '', s)
  
  # reg expr to get rid of all text enclosed by []
  s = re.sub(r'\[[^\]]*\]', '', s)
  
  # Remove extra white space while preserving new lines
  s = re.sub(r'[^\S\n]+', ' ', s) 
  
  # condense consecutive new lines into a single new line
  s = re.sub(r'\n+', '\n', s)
  
  # get rid of all instances of \'
  s = re.sub(r"\\'", "'", s)
  return s
```

```python
cleaned_lyrics = {}
for key, value in raw_lyrics.items():
  cleaned_lyrics[key] = {}  # add key (album title) to dictionary
  for k, v in value.items():
    new_k = clean_title(k[:-6]) # remove last 6 char's and all non ASCII chars
    cleaned_lyrics[key][new_k] = {}
    for a, b in v.items():
      new_lyrics = clean_string(b)
      cleaned_lyrics[key][new_k][a] = new_lyrics
        
```

```python
print(cleaned_lyrics['Encore'])
```

```python
# I will use VADER
# Considerations: VADER struggles with very long text or strings which lack clear sentence boundaries
# VADER is designed to analyze sentiment on a sentence or document level, as it utilizes context to do it's sentiment scoring
# Conclusion: Rather than a list of words, should feed Vader a list of lines or 'bars' which break each song into smaller strings

divided_lyrics = {}
for key, value in cleaned_lyrics.items():
  divided_lyrics[key] = {}  # add key (album title) to dictionary
  divided_lyrics[key]['a-sentiment'] = 0
  for k, v in value.items():
    divided_lyrics[key][k] = {}
    divided_lyrics[key][k]['s-sentiment'] = 0
    for a, b in v.items():
      split_lyrics = [line for line in b.split('\n') if line] # splits by new line, removing empty lines
      divided_lyrics[key][k][a] = split_lyrics
      if key == 'The Marshall Mathers LP':
        print(split_lyrics)
print(divided_lyrics['The Marshall Mathers LP'])
```

```python
import nltk
from nltk.sentiment import vader
nltk.download('vader_lexicon')
nltk.download('stopwords')
```

```python
sia = vader.SentimentIntensityAnalyzer()
for album, tracks in divided_lyrics.items(): # key is album, value is list of songs
  album_sentiment = 0
  for song, lines in tracks.items(): # song is keys under album, v is dictionary
    song_sentiment = 0
    if isinstance(lines, dict):
      for lyrics, l in lines.items(): # lyrics is keys under song
        if isinstance(l, list):
          for bars in l:
            song_sentiment += sia.polarity_scores(bars)['compound']
      divided_lyrics[album][song]['s-sentiment'] = song_sentiment
      album_sentiment += song_sentiment
  divided_lyrics[album]['a-sentiment'] = album_sentiment


print(divided_lyrics['Encore'])
        
```

```python
# do not care about song title, only care about the order of songs in each album
def plot_sentiment_songs(album_dict):
  album_scores = []
  for key, value in album_dict.items(): # value is dictionary under album
    song_scores = []
    plot_title = "Sentiment of " + key
    for k, v in value.items(): # v is dictionary under song
      if isinstance(v, dict): # k is name of song
        for a, b in v.items(): # b is list of songs or sentiment score
          if isinstance(b, float): # b is sentiment score
            song_scores.append((k, b))
    album_scores.append((key, song_scores))
  return album_scores
```

```python
import matplotlib.pyplot as plt

data_set = plot_sentiment_songs(divided_lyrics)
def sentiment_charter(t):
  album = t[0]
  song_name, sent_score = zip(*t[1])
  track_numbers = list(range(1, len(sent_score) + 1))
  
  colors = ['orange' if score >= 0 else 'purple' for score in sent_score]
    
  plt.bar(track_numbers, sent_score, color=colors) # changed from song name to track number, as names were too long
  plt.title(album)
  plt.xlabel('Track Number')
  plt.ylabel('Sentiment')
  plt.xticks(track_numbers)
```
